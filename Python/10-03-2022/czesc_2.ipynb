## Przygotowywanie danych (zbiór treningowy i testowy)

```python
from sklearn.model_selection import train_test_split
```

```python
y = dataset.pop("Temperature (C)")
dataset.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, y, test_size=0.2)
y_test = list(y_test)
X_test
```

## Funkcje miar jakości modeli

```python
def get_mse(y_test: list[float], y_predicted: list[float]) -> float:
    mse = 0
    for index in range(len(y_test)):
        y = y_test[index]
        y_pred = round(float(y_predicted[index]), 2)
        residual = round(y - y_pred, 2)

        mse += residual**2

    mse /= len(y_test)
    return mse


def get_rmse(mse: float) -> float:
    rmse = mse ** (1/2)
    return rmse


def get_mae(y_test: list[float], y_predicted: list[float]) -> float:
    mae = 0
    for index in range(len(y_test)):
        y = y_test[index]
        y_pred = float(y_predicted[index])
        residual = y - y_pred

        mae += abs(residual)

    mae /= len(y_test)
    return mae


def get_r2(y_test: list[float], y_predicted: list[float]) -> float:
    y_avg = sum(y_test)/len(y_test)
    numerator = 0
    denominator = 0
    for index in range(len(y_test)):
        y = y_test[index]
        y_pred = float(y_predicted[index])

        numerator += (y_pred - y_avg)**2
        denominator += (y - y_avg)**2

    r2 = numerator / denominator
    return r2
```

# 1. Regresja liniowa

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
y_predicted = regression.predict(X_test)
```

```python
mse = get_mse(y_test, y_predicted)
rmse = get_rmse(mse)
mae = get_mae(y_test, y_predicted)
r2 = get_r2(y_test, y_predicted)

print("MSE =", mse)
print("RMSE =", rmse)
print("MAE =", mae)
print("R2 =", r2)
```
